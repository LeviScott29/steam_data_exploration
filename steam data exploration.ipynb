```python
import pandas as pd
```

```python

path = r"C:\Users\angel\Downloads\archive (6)\steamspy_data.csv"
df = pd.read_csv(path)
pd.set_option("display.max_columns", None)
```

```python
df.info()
```

```python
# drop uneccesary columns
df.drop(columns=['score_rank', 'initialprice', 'discount', 'positive', 'negative', 'ccu', 'owners', 'tags'], inplace=True)
df.head()
```

```python
df.describe()
```

```python
#change column names to make more sense 
df.rename(columns={'average_forever': 'average_play_forever', 'average_2weeks': 'average_play_2weeks', 'median_forever': 'median_play_forever', 'median_2weeks':'median_play_2weeks'}, inplace=True)
#adjust price column from cents to dollars
df['price'] = df['price']/100
```

```python
#counts for most common developers, publishers, and genres 
print(df['developer'].value_counts())
print('\n')
print(df['publisher'].value_counts())
print('\n')
print(df['genre'].value_counts())
```

```python
#find nulls in data
df.isnull().sum()
```

```python
#clean out nulls
cleaned_df= df.dropna()
cleaned_df.isnull().sum()
```

```python
#find games with highest total average play time 
highest_playtime_alltime_df = cleaned_df.sort_values(by='average_play_forever', ascending = False)
highest_playtime_alltime_df.head()
```

```python
# find highest average played game by playtime over 2 weeks
highest_playtime_2weeks_df= cleaned_df.sort_values(by='average_play_2weeks', ascending= False)
highest_playtime_2weeks_df.head()
```
